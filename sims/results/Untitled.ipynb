```python
import os
```

```python
files = [x for x in os.listdir('./') if x.endswith('.csv')]
```

```python
for f in files:
    tdf = pd.read_csv(f)
    tdf['lambda'].hist(normed=True)
    plt.vlines(tdf['lambda'].mean(), 0,1)
    print(f)
    print(tdf['lambda'].mean(), tdf['lambda'].var())
    plt.show()
    plt.close()
    
```
