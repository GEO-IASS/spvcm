```python
sb.set_context('talk')
fbetas = np.vstack(q.trace.Stochastics['betas'])
truebetas = [-3, 6, 0, 2, 5, -4]
N,p = fbetas.shape
for i in range(p):
    plt.plot(fbetas.T[i])
    plt.title("$\\beta_{}$".format(i))
    print('Mean: {}, True: {} Variance: {}'.format(
            np.mean(fbetas.T[i]), truebetas[i], np.var(fbetas.T[i])))
    plt.show()
```

```python
sb.set_context('talk')
fthetas = np.hstack(q.trace.Stochastics['thetas'])
N,p = fthetas.T.shape
for i in range(p):
    plt.plot(fthetas[i])
    plt.title("$\\theta_{}$".format(i))
    print('Mean: {}, Variance: {}'.format(
            np.mean(fthetas[i]), np.var(fthetas[i])))
    plt.show()
```
